# PIE CHART

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index(name='Rating Count')
```

```python
df_ratings = final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index(name='Rating Count')
```

```python
df_ratings
```

```python
# OBSERVATION
When rating is between (4.5 - 4.9) it is #EXCELLENT
When rating is between (4.0 - 4.4) it is #VERY GOOD
When rating is between (3.5 - 3.9) it is #GOOD
When rating is between (3.4 - 2.5) it is #AVERAGE
When rating is between (2.4 - 1.8) it is #POOR
When rating is (0.0) it is #NOT RATED
```

```python
# SHOWING RELATION BETWEEN AGGREGATE RATING AND RATING COUNT
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=df_ratings,hue='Rating color',palette=['pink','Red','Orange','Yellow','Green','Green'])
plt.show()
```

```python
#OBSERVATION
1) NOT RATED IS HIGH
2) MAXIMUM NO. OF RATINGS IS BETWEEN (2.5 - 3.4)
```

```python
#COUNT PLOT
sns.countplot(x="Rating color",data=df_ratings,hue='Rating color',palette=['pink','Red','Orange','Yellow','Green','Green'])
plt.show()
```
